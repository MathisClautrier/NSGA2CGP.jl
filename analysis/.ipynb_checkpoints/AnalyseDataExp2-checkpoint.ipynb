```python
import pandas as pd
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../logs/2020-10-16T14:35:36.962.csv",names=["Date","Pkg","Info","Num","Max","Mean","Std"])
```

```python
m=data.shape[0]
```

```python
cent=[[],[],[]]
```

```python
for i in range(m):
    x=data[data["Num"]==i+1].values
    cent[0].append(x[0][4])
    cent[1].append(x[0][5])
    cent[2].append(x[0][6])
    
```

```python
len(cent[0])
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2,figsize=(15,5))
ax[0].plot(cent[0],'b.')
ax[1].plot(cent[1],'b.')
ax[0].set_title("Max")
ax[1].set_title("Mean")
```
