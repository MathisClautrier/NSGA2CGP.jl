```python
import pandas as pd
%matplotlib inline 
import matplotlib.pyplot as plt
```

Ici on regarde l'effet de la correction d'une erreur d'incrémentation qui effacait le premier individu au lieu de le sélectionner. On le compare enfin à la même expérience sans cette correction. Cette comparaison est "injuste", l'expérience avec correction est fait sur 15 individus alors que celle avec est fait sur 5.

```python
data = pd.read_csv("../logs/2020-10-28T15:59:19.257.csv",names=["Date","Pkg","Info","Num","Max","Mean","Std"])
```

```python
m=data.shape[0] LLL
```

```python
cent=[[],[],[]]
fros=[[],[],[]]
pacm=[[],[],[]]
```

```python
for i in range(m//3):
    x=data[data["Num"]==i+1].values
    cent[0].append(x[0][4])
    cent[1].append(x[0][5])
    cent[2].append(x[0][6])
    fros[0].append(x[1][4])
    fros[1].append(x[1][5])
    fros[2].append(x[1][6])
    pacm[0].append(x[2][4])
    pacm[1].append(x[2][5])
    pacm[2].append(x[2][6])
    
```

```python
fig, ax = plt.subplots(nrows=3,ncols=2,figsize=(10,10))
ax[0,0].plot(cent[0],'b.')
ax[0,1].plot(cent[1],'b.')
ax[1,0].plot(fros[0],'b.')
ax[1,1].plot(fros[1],'b.')
ax[2,0].plot(pacm[0],'b.')
ax[2,1].plot(pacm[1],'b.')
ax[0,0].set_title("Max")
ax[0,1].set_title("Mean")
```

L'apprentissage sur Frostbite (deuxiéme ligne) reste difficile avec un score statique. Ce n'est pas surprennant Frostbite utilise 85% du graphe TPG. Aprés l'implémentation de la crowding distance, l'augmentation du nombre de colonne a de grande chance d'améliorer l'apprentissage sur Frostbite. De plus, il n'y a toujours pas sauvegarde des élites sur chaque jee d'une génération à une autre ce que l'on cherche. Probléme à solutionner.

```python
data2 = pd.read_csv("../logs/2020-10-16T13:48:00.924.csv",names=["Date","Pkg","Info","Num","Max","Mean","Std"])
m2=data2.shape[0]
cent2=[[],[],[]]
fros2=[[],[],[]]
pacm2=[[],[],[]]
for i in range(m2//3):
    x=data2[data2["Num"]==i+1].values
    cent2[0].append(x[0][4])
    cent2[1].append(x[0][5])
    cent2[2].append(x[0][6])
    fros2[0].append(x[1][4])
    fros2[1].append(x[1][5])
    fros2[2].append(x[1][6])
    pacm2[0].append(x[2][4])
    pacm2[1].append(x[2][5])
    pacm2[2].append(x[2][6])

fig, ax = plt.subplots(nrows=3,ncols=2,figsize=(10,10))
ax[0,0].plot(cent[0],'b.')
ax[0,1].plot(cent[1],'b.')
ax[1,0].plot(fros[0],'b.')
ax[1,1].plot(fros[1],'b.')
ax[2,0].plot(pacm[0],'b.')
ax[2,1].plot(pacm[1],'b.')
ax[0,0].plot(cent2[0],'r.')
ax[0,1].plot(cent2[1],'r.')
ax[1,0].plot(fros2[0],'r.')
ax[1,1].plot(fros2[1],'r.')
ax[2,0].plot(pacm2[0],'r.')
ax[2,1].plot(pacm2[1],'r.')
ax[0,0].set_title("Max")
ax[0,1].set_title("Mean")
```

Même si cette comparaison est "injuste" il semble que cette correction à grandement amélioré l'apprentissage. Il faudra refaire l'expérience avec les mêmes paramétres pour en être certain. Cependant, çà n'a pas beaucoup d'intérêt sur l'objectif de ces expériences.
